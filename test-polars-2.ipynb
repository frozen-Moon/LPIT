https://docs.pola.rs/user-guide/concepts/lazy-api/  
https://github.com/pola-rs/polars/blob/main/docs/assets/data/iris.csv

```python
# Eager API
# - Every step is executed immediately returning the intermediate results
df = pl.read_csv("iris.csv")
df_small = df.filter(pl.col("sepal_length") > 5)
df_agg = df_small.group_by("species").agg(pl.col("sepal_width").mean())
print(df_agg)
```

```python
# Lazy API 
# - Define all the steps first, then execute the query
# - The query planner can perform optimizations
#   (e.g. read only rows with sepal length greater than 5)
# - The query is only evaluated when it is collected
q = (
    pl.scan_csv("iris.csv")
    .filter(pl.col("sepal_length") > 5)
    .group_by("species")
    .agg(pl.col("sepal_width").mean())
)
```

```python
print(q.explain())
```

```python
df_agg = q.collect()
```

```python
print(df_agg)
```
