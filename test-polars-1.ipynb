```python
import polars as pl
from datetime import date
```

```python
# Dataframe example
df = pl.DataFrame(
    {
        "name": ["Dean Lamb", "Dave Mustaine", "Chuck Schuldiner", 
                 "Adam De Micco", "Marty Friedman"],
        "birthdate": [
            date(1987, 11, 16),
            date(1961, 9, 13),
            date(1967, 5, 13),
            date(1988, 3, 16),
            date(1962, 12, 8),
        ],
        "weight": [57.9, 72.5, 53.6, 83.1, 62.3],  # (kg)
        "height": [1.56, 1.77, 1.65, 1.75, 1.70],  # (m)
    }
)
```

```python
df
```

```python
print(df)
```

```python
df.write_csv("output.csv")
```

```python
df_csv = pl.read_csv("output.csv", try_parse_dates=True)
```

```python
print(df_csv)
```

```python
# select 
# - Creates a new dataframe by selecting and calculating columns
# - Each expression will map to a column in the result dataframe
result = df.select(
    pl.col("name"),
    pl.col("birthdate").dt.year().alias("birth_year"),
    (pl.col("weight") / (pl.col("height") ** 2)).alias("bmi"),
)
```

```python
print(result)
```

```python
# Expression expansion 
result = df.select(
    pl.col("name"),
    (pl.col("weight", "height") * 0.95).round(2).name.suffix("-5%"),
)
```

```python
print(result)
```

```python
print(df)
```

```python
result = df.with_columns(
    birth_year=pl.col("birthdate").dt.year(),
    bmi=pl.col("weight") / (pl.col("height") ** 2),
)
```

```python
print(result)
```

```python
# with_columns 
# - Creates a new dataframe that contains columns from the original dataframe 
#   along with new columns according to the input expressions
result = df.with_columns(
    pl.col("birthdate").dt.year().alias("birth_year "),
    (pl.col("weight") / (pl.col("height") ** 2)).alias("bmi"),
)
```

```python
print(result)
```

```python
# filter
# - Filters rows based on one or more expressions that evaluate to Boolean
result = df.filter(pl.col("birthdate").dt.year() != 1985)
```

```python
print(df)
print(result)
```

```python
# Multiple expressions as separate parameters to create composite conditions
result = df.filter(
    pl.col("birthdate").is_between(date(1982, 12, 31), date(1996, 1, 1)),
    pl.col("height") > 1.7,
)
```

```python
print(result)
```

```python
# group_by
# -  Rows are grouped according to the unique values of the expressions
# -  e.g. Count how many people were born in each decade
result = df.group_by(
    (pl.col("birthdate").dt.year() // 10 * 10).alias("decade"),
    maintain_order=True,
).len()
```

```python
print(result)
```

```python
result = df.group_by(
    (pl.col("birthdate").dt.year() // 10 * 10).alias("decade"),
    maintain_order=True,
).agg(
    pl.len().alias("sample_size"),
    pl.col("weight").mean().round(2).alias("avg_weight"),
    pl.col("height").max().alias("tallest"),
)
```

```python
print(result)
```
