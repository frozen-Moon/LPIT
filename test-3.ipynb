```python
class Vehiculo:

    # Atributo de clase. Compartido por todas las instancias.
    tipo_vehiculo = "Transporte"

    # Constructor de la clase. Se ejecuta automáticamente al crear una instancia.
    def __init__(self, marca, modelo, velocidad=0):
        self.marca = marca          # Atributo de instancia
        self.modelo = modelo        # Atributo de instancia
        self.velocidad = velocidad  # Atributo de instancia

    # Métodos de instancia. Operan sobre los datos de una instancia específica (self).

    # Incrementar la velocidad del vehículo
    def acelerar(self, incremento):
        self.velocidad += incremento
        return f"{self.marca} {self.modelo} ahora va a {self.velocidad} km/h"

    # Reducir la velocidad del vehículo
    def frenar(self, decremento):
        self.velocidad = max(0, self.velocidad - decremento)
        return f"{self.marca} {self.modelo} ahora va a {self.velocidad} km/h"

    # Métodos de clase. Recibem la clase como argumento (cls).
    # Se usan para acceder o modificar los atributos de clase.
    @classmethod
    def descripcion_tipo(cls):
        return f"Todos los vehículos son de tipo: {cls.tipo_vehiculo}"

    # Método estático. No depende de la instancia ni de la clase.
    # Se comporta como una función normal, pero pertenece a la clase.
    @staticmethod
    def es_vehiculo_rapido(velocidad):
        return velocidad > 120
```

```python
# ==========================
# EJEMPLO DE USO (INSTANCIAS)
# ==========================

# Crear dos instancias de la clase Vehiculo
coche1 = Vehiculo("Toyota", "Corolla")
coche2 = Vehiculo("Tesla", "Model 3", 50)
```

```python
print(coche1.marca, '-', coche1.modelo)
```

```python
print(coche2.marca, '-', coche2.modelo)
```

```python
print(coche1.acelerar(30))
```

```python
print(coche2.frenar(10))
```

```python
print(Vehiculo.descripcion_tipo())
```

```python
print(Vehiculo.es_vehiculo_rapido(130))
```
