```python
import polars as pl
from datetime import date
```

```python
# Dataframe example
df = pl.DataFrame(
    {
        "name": ["Dean Lamb", "Dave Mustaine", "Chuck Schuldiner", 
                 "Adam De Micco", "Marty Friedman"],
        "birthdate": [
            date(1987, 11, 16),
            date(1961, 9, 13),
            date(1967, 5, 13),
            date(1988, 3, 16),
            date(1962, 12, 8),
        ],
        "weight": [57.9, 72.5, 53.6, 83.1, 62.3],  # (kg)
        "height": [1.56, 1.77, 1.65, 1.75, 1.70],  # (m)
    }
)
```

```python
print(df)
```

```python
df.write_parquet("data.parquet")
```

```python
df2 = pl.read_parquet("data.parquet")
```
